```python
#Correlation Heatmap For Numeric Feature
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(test_df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

#Insights

- LoanAmount shows a positive correlation with ApplicantIncome, indicating  higher-income applicants tend to request larger loan amounts.

- DTI is positively correlated with LoanAmount and negatively correlated with ApplicantIncome, confirming that higher loans relative to income increase repayment burden.

- Credit_History has weak correlation with most numeric variables, suggesting it acts as an independent approval factor rather than being income-driven.

- CoapplicantIncome has low to moderate correlation with LoanAmount, implying secondary income provides limited influence compared to primary income.

#Insights

- Applicants from rural areas with lower income levels and higher
debt-to-income ratios represent a higher credit risk. A DTI above 40%
- indicates increased repayment burden, particularly for applicants with
limited income capacity.


- Applicants from rural property areas with lower applicant income and higher loan-to-income ratios represent the high-risk segment.
-  These applicants typically show greater financial stress, especially when credit history is weak, indicating a higher likelihood of repayment challenges. -  Setting stricter approval thresholds for high debt-to-income ratios can help reduce lending risk.
