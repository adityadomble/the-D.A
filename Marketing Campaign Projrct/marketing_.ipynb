#Insight

```python
#visualize Revenue by Channel
```

```python
plt.figure(figsize= (8,5))
sns.barplot(x=revenue_channel.index, y=revenue_channel.values)

plt.title("Revenue by Marketing Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("total_revenue")

plt.show()
```

#Insights

 - We can clearly see google ads marketing channel with $ 32million dollars
    on second place influencer channel with $ 33 million
 - Clear from the bar chart that the revenues generated from different marketing channels have been represented. Google Ads is the channel that generates the maximum revenue, followed by Influencer and Email marketing. Affiliate generates the least revenues among all the channels.

```python
#MonthWise Revenue
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python

revenue_trend = df.groupby("Date")["Revenue"].sum()

revenue_trend.plot()

plt.title("Revenue Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")

plt.show()
```

#Insight

-   Festivals or holidays

-   Discount campaigns

-   Increased customer demand 📈



Low-performing months may indicate:

- Weak campaigns
- Low market demand

 -  The analysis of revenues in each month can help in understanding the trends in marketing performance. There are some months in which revenues are higher, and this indicates the peak period in which marketing campaigns are working effectively.
