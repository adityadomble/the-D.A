```python
#.Sales by Region
```

```python
region=df.groupby("Region")["Sales"].sum().sort_values(ascending = False)
```

```python
#Correlation Heatmap
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df[["Sales","Profit","Quantity","Discount"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```
